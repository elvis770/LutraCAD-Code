EXTRACTING MAJOR PARAMETER DATA FROM THE STL FILES

```python
import os
import pandas as pd
from stl import mesh
import numpy as np

# Directory containing the folders
base_dir = '/Users/elvisechefu/Desktop/LutraCAD2/Anonymous_20240507025011'

# Initialize an empty list to store the data
data = []

# Loop through each folder
for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)
    if os.path.isdir(folder_path):
        stl_files = [f for f in os.listdir(folder_path) if f.endswith('.stl') and f != 'library.stl']  # Exclude 'library.stl'
        for stl_file in stl_files:
            stl_file_path = os.path.join(folder_path, stl_file)
            stl_mesh = mesh.Mesh.from_file(stl_file_path)
            bounding_box_dimensions = stl_mesh.points.ptp(axis=0)
            volume, cog, inertia = stl_mesh.get_mass_properties()
            surface_area = sum(np.linalg.norm(np.cross(v0 - v1, v0 - v2)) / 2 for v0, v1, v2 in stl_mesh.vectors)
            data.append([os.path.basename(stl_file_path), os.path.getsize(stl_file_path), len(stl_mesh.vectors), bounding_box_dimensions, surface_area, volume, cog, 'Material', bounding_box_dimensions[0], bounding_box_dimensions[1]])

# Create a DataFrame and save to CSV
columns = ['File Name', 'File Size', 'Triangles', 'Bounding Box Dimensions', 'Surface Area', 'Volume', 'Center of Mass', 'Material', 'Length', 'Width']
df2 = pd.DataFrame(data, columns=columns)
df2.to_csv('stl_metadata.csv', index=False)
```

```python
import os
import json

def parse_actions_dat(actions_dat_file_path):
    parsed_data = {
        "folder_name": os.path.basename(os.path.dirname(actions_dat_file_path)),
        "HeelLift": None,
        "InsoleLength": None,
        "InsoleHeight": None,
        "TransformationData": None,
        "Parameters": {
            "SizeTable": None,
            "LengthInMM": None,
            "WidthInMM": None,
            "HeightInMM": None,
            "PatchSmooth": None,
            "PatchOffset": None,
            "SmoothFactor": None,
            "EnableFlatten": None,
            "Thickness": None
        }
    }

    try:
        with open(actions_dat_file_path, 'r') as file:
            actions_dat_data = json.load(file)

            for item in actions_dat_data["List"]:
                # Parse HeelLift data
                if item.get("Type") == "HeelLift":
                    heel_lift_data = {
                        "Height": item.get("Height"),
                        "Length": item.get("Length"),
                        "Side": item.get("Side")
                    }
                    parsed_data["HeelLift"] = heel_lift_data

                # Parse InsoleLength data
                elif item.get("Type") == "InsoleLength":
                    insole_length_data = {
                        "Length": item.get("Length"),
                        "Width": item.get("Width"),
                        "Height": item.get("Height"),
                        "Resources": item.get("Resources")
                    }
                    parsed_data["InsoleLength"] = insole_length_data

                # Parse InsoleHeight data
                elif item.get("Type") == "InsoleHeight":
                    insole_height_data = {
                        "Height": item.get("Height"),
                        "Resources": item.get("Resources")
                    }
                    parsed_data["InsoleHeight"] = insole_height_data

                # Parse Transformation data
                elif item.get("LeftTransform") or item.get("RightTransform"):
                    transformation_data = {
                        "LeftTransform": item.get("LeftTransform"),
                        "RightTransform": item.get("RightTransform"),
                        "Resources": item.get("Resources")
                    }
                    parsed_data["TransformationData"] = transformation_data

                # Parse Parameters data
                parameters = ["SizeTable", "LengthInMM", "WidthInMM", "HeightInMM", "PatchSmooth", "PatchOffset", "SmoothFactor", "EnableFlatten", "Thickness"]
                for param in parameters:
                    if item.get(param) is not None:
                        parsed_data["Parameters"][param] = item.get(param)

    except Exception as e:
        print(f"Error parsing actions.dat file: {e}")

    return parsed_data

# Define the directory containing all the folders
root_directory = "/Users/elvisechefu/Desktop/LutraCAD2/Anonymous_20240507025011"

# Collect all parsed data
all_data = []

# Iterate through each folder in the root directory
for folder_name in os.listdir(root_directory):
    # Construct the full path of the folder
    folder_path = os.path.join(root_directory, folder_name)
    
    # Check if it's a directory
    if os.path.isdir(folder_path):
        #print("Folder:", folder_name)
        
        # Search for actions.dat files within the folder
        for file_name in os.listdir(folder_path):
            if file_name == "actions.dat":
                actions_dat_file_path = os.path.join(folder_path, file_name)
                #print("  Actions.dat file found:", actions_dat_file_path)
                
                # Parse the actions.dat file
                parsed_data = parse_actions_dat(actions_dat_file_path)
                
                # Append parsed data to all_data list
                all_data.append(parsed_data)

# Display collected data in the desired format
for data in all_data:
    print("\nParameters:")
    print(f"SizeTable: {data['Parameters']['SizeTable']}")
    print(f"LengthInMM: {data['Parameters']['LengthInMM']}")
    print(f"WidthInMM: {data['Parameters']['WidthInMM']}")
    print(f"HeightInMM: {data['Parameters']['HeightInMM']}")
    print(f"PatchSmooth: {data['Parameters']['PatchSmooth']}")
    print(f"PatchOffset: {data['Parameters']['PatchOffset']}")
    print(f"SmoothFactor: {data['Parameters']['SmoothFactor']}")
    print(f"EnableFlatten: {data['Parameters']['EnableFlatten']}")
    print(f"Thickness: {data['Parameters']['Thickness']}")
    print(f"Folder: {data['folder_name']}")
```
